```python
# Title your chart "Big O Notation"
#Label your x axis "Elements"
#Label your y axis "Operations"
#Label your curves or make a legend for the curves
#Use LaTex notation where possible

import math
import matplotlib.pyplot as plt

x = range(1, 20)
# O(1)
y1 = [1 for n in x]

y2 = [math.log(n) for n in x]

y3 = x

y4 = [n * math.log(n) for n in x]

y5 = [n**2 for n in x]

y6 = [2**n for n in x]

y7 = [math.factorial(n) for n in x]

y8 = [n**n for n in x]
```

```python
plt.figure(figsize = (16, 8))

plt.title('Big O Notation', fontsize = 20)

plt.xlabel('Elements', fontsize=14)
plt.ylabel('Operations', fontsize=14)

# O(1)
plt.plot(x, y1, label='$\mathcal{O}(1)$', color='purple')

# O(log n)
plt.plot(x, y2, label='$\mathcal{O}(logn)$', color='black')

# O(n)
plt.plot(x, y3, label='$\mathcal{O}(n)$', color='green')

# O(n log n)
plt.plot(x, y4, label='$\mathcal{O}(nlogn)$', color='yellow')

# O(n^{2})
plt.plot(x, y5, label='$\mathcal{O}(n^{2})$', color='orange')

# O(2^{n})
plt.plot(x, y6, label='$\mathcal{O}(2^{n})$', color='brown')

# O(n!)
plt.plot(x, y7, label='$\mathcal{O}(n!)$', color='blue')

# O(n^{n})
plt.plot(x, y8, label='$\mathcal{O}(n^{n})$', color='red')

plt.xlim(1, 11)
plt.ylim(0, 50)

plt.legend(loc='upper right', fontsize=14)

plt.show()
```
