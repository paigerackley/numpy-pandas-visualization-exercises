```python
# 1. Copy the users and roles DataFrames from the examples above.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})

users
```

```python
roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
# 2. What is the result of using a right join on the DataFrames?
pd.merge(users, roles, left_on='role_id', right_on='id', how='right')
# Columns with matching names are renamed with _x or _y depending on which datafram
# All rows on right kept
# IF row on left matches row on right = combined
# if row on left does not match on right = discarded
# if row on right does not match on left = new row is added
# rows = match if left and right match info(all info)
```

```python
# 3. What is the result of using an outer join on the DataFrames?
pd.merge(users, roles, left_on='role_id', right_on='id', how='outer')
# matching names = _x or _y depending on which dataframe
# ALL ROWS kept = rows that match = combined , rows not match = new row
```

```python
# 12. Create a column named mileage_difference like you did in the DataFrames exercises; this column should 
# contain the difference between highway and city mileage for each car.
mpg['mileage_difference'] = mpg.highway - mpg.city
mpg.head()
```

```python
# 13. Create a column named average_mileage like you did in the DataFrames exercises; this is the mean of the 
# city and highway mileage.
mpg['average_mileage'] = (mpg.highway + mpg.city) / 2
mpg.head()
```

```python
# 14. Create a new column on the mpg dataset named is_automatic that holds boolean values denoting whether the 
# car has an automatic transmission.
mpg['is_automatic'] = mpg.trans.str.contains('auto')
mpg.head()
# OR
# mpg['is_automatic'] = mpg.trans.apply(lambda value : value[0] == 'a')
```

```python
# 15. Using the mpg dataset, find out which which manufacturer has the best miles per gallon on average?
mpg.groupby('manufacturer').average_mileage.mean().sort_values().tail(1)
```

```python
# 16. Do automatic or manual cars have better miles per gallon?
mpg.groupby('is_automatic').average_mileage.agg('mean')
```

```python
# 7. Write the code necessary to create a cross tabulation of the 
# number of titles by department. (Hint: this will involve a 
# combination of SQL code to pull the necessary data and python/pandas 
# code to perform the manipulations.)
url = get_db_url("employees")

query = """
SELECT * FROM titles
"""

titles = pd.read_sql(query, url)
titles.head(5)
```

```python
url = get_db_url("employees")

query = """
SELECT * FROM departments 
"""

departments = pd.read_sql(query, url)
departments.head(5)
```

```python
url = get_db_url("employees")

query = """
SELECT * FROM dept_emp 
"""

dept_emp = pd.read_sql(query, url)
dept_emp.head(5)
```

```python
# employees_dept = pd.concat([titles, departments, dept_emp], axis=1)
employees_dept = pd.merge(dept_emp, departments, left_on='dept_no', right_on='dept_no', how='inner')
employees_dept
```

```python
employees_dt = pd.merge(employees_dept, titles, left_on='emp_no', right_on='emp_no', how='inner')
employees_dt
```

```python
pd.crosstab(employees_dt.title, employees_dt.dept_name)
```
